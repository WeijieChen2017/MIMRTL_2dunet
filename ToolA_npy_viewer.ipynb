## channel last

```python
plt.figure(figsize=(6,2), dpi=200)

plt.subplot(131)
plt.imshow(data[:, :, 0], cmap='gray')
plt.axis("off")

plt.subplot(132)
plt.imshow(data[:, :, 1], cmap='gray')
plt.axis("off")

plt.subplot(133)
plt.imshow(data[:, :, 2], cmap='gray')
plt.axis("off")
```
